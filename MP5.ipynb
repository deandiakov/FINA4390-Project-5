```python
import pandas as pd

dataA = pd.read_csv('data/Cleaned_LoanStats3a_securev1.csv', header='infer')
dataB = pd.read_csv('data/Cleaned_LoanStats3b_securev1.csv', header='infer')
dataC = pd.read_csv('data/Cleaned_LoanStats3c_securev1.csv', header='infer')
dataD = pd.read_csv('data/Cleaned_LoanStats3d_securev1.csv', header='infer')

dataA
```

```python
all_data = pd.concat([dataA, dataB, dataC, dataD])
all_data.describe()
```

```python
# Convert loan status to number
loan_vals = set()

for ind, row in all_data.iterrows():
    loan_vals.add(row['loan_status'])
```

```python
print(loan_vals)
```

```python
ineligible_vals = ['Default', 'Does not meet the credit policy. Status:Charged Off', 'Charged Off']

res = []
for ind, row in all_data.iterrows():
    if row['loan_status'] in ineligible_vals:
        res.append(1)
    else:
        res.append(0)

all_data['loan_status_bin'] = res
```

```python
# Variables that correlate with PRODUSE_3

import seaborn as sb
import matplotlib.pyplot as plt0

# figure size
plt.figure(figsize=(1, 10))

benefits = all_data.corr()['loan_status_bin'].to_frame().sort_values('loan_status_bin', ascending = False).transpose()
filter = abs(benefits) > 0.12
benefits = benefits[filter]
benefits = benefits.dropna(axis = 1).transpose()
sb.heatmap(benefits, cmap="YlGnBu")

# Axes
plt.yticks(rotation=0) # This sets the yticks "upright" with 0, as opposed to sideways with 90.
plt.show()
```

```python
# Feature Selection
import numpy as np
selected_cols = ['annual_inc', 'loan_status_bin', 'last_fico_range_low', 'last_fico_range_high', 
                 'collections_12_mths_ex_med', 'delinq_2yrs', 'emp_length', 'home_ownership', 'loan_amnt',
                 'open_acc', 'pub_rec', 'purpose', 'revol_bal', 'grade']

data = all_data[selected_cols]
data = data.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
# Process Data
predictor = data.iloc[:,data.columns!='loan_status_bin']
target = data.iloc[:,data.columns=='loan_status_bin']

predictor.head()
```

```python
# Dummify data

categorical_variables = [col for col in predictor.columns.values if predictor[col].dtype=='object']

df_categorical = predictor[categorical_variables]

df_numeric = predictor.drop(categorical_variables, axis=1)
```

```python
dummy_code_cat_vars = pd.get_dummies(df_categorical, drop_first=True)
```

```python
dummy_code_cat_vars.head()
```

```python
df_predictor = pd.concat([df_numeric, dummy_code_cat_vars], axis=1)
df_predictor.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(df_predictor, target, test_size = 0.30, random_state=0,stratify=target)
print("x_train ",X_train.shape)
print("x_test ",X_test.shape)
print("y_train ",y_train.shape)
print("y_test ",y_test.shape)

from sklearn.preprocessing import StandardScaler
init_scaler = StandardScaler()

X_train_scaled = pd.DataFrame(init_scaler.fit_transform(X_train))
X_test_scaled = pd.DataFrame(init_scaler.transform(X_test))

X_train_scaled.columns = X_train.columns.values
X_test_scaled.columns = X_test.columns.values
X_train_scaled.index = X_train.index.values
X_test_scaled.index = X_test.index.values 

X_train = X_train_scaled
X_test = X_test_scaled
X_train.describe()
```

```python
# ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import regularizers
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, mean_squared_error, mean_absolute_error, r2_score, roc_curve, roc_auc_score
import seaborn as sns 
import numpy as np

def plot_loss_accuracy(history):
    historydf = pd.DataFrame(history.history, index=history.epoch)
    plt.figure(figsize=(8, 6))
    historydf.plot(ylim=(0, max(1, historydf.values.max())))
    loss = history.history['loss'][-1]
    acc = history.history['accuracy'][-1]
    plt.title('Loss: %.3f, Accuracy: %.3f' % (loss, acc))

def plot_loss(history):
    historydf = pd.DataFrame(history.history, index=history.epoch)
    plt.figure(figsize=(8, 6))
    historydf.plot(ylim=(0, historydf.values.max()))
    plt.title('Loss: %.3f' % history.history['loss'][-1])
    
def plot_confusion_matrix(model, X, y):
    y_pred = model.predict_classes(X, verbose=0)
    plt.figure(figsize=(8, 6))
    sns.heatmap(pd.DataFrame(confusion_matrix(y, y_pred)), annot=True, fmt='d', cmap='YlGnBu', alpha=0.8, vmin=0)
    plt.title('Confusion matrix of the classifier')
    plt.xlabel('Predicted')
    plt.ylabel('True')

def create_model(first_layer, second_third_layer, fourth_fifth_layer, activation):    
    model = Sequential()
    model.add(Dense(first_layer, input_shape=(43,), activation=activation))
    model.add(Dense(second_third_layer, activation=activation))
    model.add(Dense(second_third_layer, activation=activation))
    model.add(Dense(fourth_fifth_layer, activation=activation))
    model.add(Dense(fourth_fifth_layer, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=Adam(lr=.01), loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# First attempt, without hyperparameter tuning
naive_model = create_model(12, 36, 24, 'relu')

# Early stopping function
early_stop = EarlyStopping(monitor='val_loss', patience=10, verbose=1) 

# Class weights based on data (ratio of 0s to 1s)
class_weight = {0: 1,
                1: 14}

ann = naive_model.fit(x=X_train, y=y_train, verbose=0, epochs=100, callbacks=[early_stop], validation_split=.3, class_weight=class_weight)
naive_model.summary()

# Metrics
plot_loss(ann)
```

```python
y_pred_proba = naive_model.predict(X_test).ravel()
# print(y_pred_proba)

fpt, tpr, t = roc_curve(y_test, y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpt, tpr, label='AUC='+str(auc))
plt.legend()
plt.show()

y_pred_ann = naive_model.predict_classes(X_test)
print(classification_report(y_test, y_pred_ann))

plot_confusion_matrix(naive_model, X_test, y_test)
```

```python
# HP Tuning ANN
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV
    
model = KerasClassifier(build_fn=create_model, epochs=100, batch_size=10, verbose=0)

# Define the RS hyperparameter distribution
p = {
    'activation': ['relu', 'elu'],
    'first_layer': [10, 12, 14],
    'second_third_layer': [32, 36, 40],
    'fourth_fifth_layer': [20, 24, 28]
}

rs_model = RandomizedSearchCV(estimator=model, param_distributions=p, n_iter=5, cv=3, random_state=111, verbose=1)
rs_output = rs_model.fit(X_train, y_train)

print('Best score is: {} using {}'.format(rs_output.best_score_, rs_output.best_params_))
```

```python
params = rs_output.best_params_
model = create_model(params['first_layer'], params['second_third_layer'], params['fourth_fifth_layer'], params['activation'])

class_weight = {0: 1.,
                1: 14.}
ann = model.fit(x=X_train, y=y_train, verbose=0, epochs=100, callbacks=[early_stop], validation_split=.3, class_weight=class_weight)
model.summary()

# Metrics
plot_loss(ann)
```

```python
y_pred_proba = model.predict(X_test).ravel()
# print(y_pred_proba)

fpt, tpr, t = roc_curve(y_test, y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpt, tpr, label='AUC='+str(auc))
plt.legend()
plt.show()

y_pred_ann = model.predict_classes(X_test)
print(classification_report(y_test, y_pred_ann))

plot_confusion_matrix(model, X_test, y_test)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(df_predictor, target, test_size = 0.30, random_state=0,stratify=target)
print("x_train ",X_train.shape)
print("x_test ",X_test.shape)
print("y_train ",y_train.shape)
print("y_test ",y_test.shape)

from sklearn.preprocessing import StandardScaler
fico_scaler = StandardScaler()

X_train_scaled = pd.DataFrame(fico_scaler.fit_transform(X_train))
X_test_scaled = pd.DataFrame(fico_scaler.transform(X_test))

X_train_scaled.columns = X_train.columns.values
X_test_scaled.columns = X_test.columns.values
X_train_scaled.index = X_train.index.values
X_test_scaled.index = X_test.index.values 

X_train = X_train_scaled
X_test = X_test_scaled
X_train.describe()
```

```python
regex = re.compile(r"\[|\]|<", re.IGNORECASE)

X_train.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_train.columns.values]


model_XG = XGBClassifier()
xg_classifier = model_XG.fit(X_train, y_train)
```
